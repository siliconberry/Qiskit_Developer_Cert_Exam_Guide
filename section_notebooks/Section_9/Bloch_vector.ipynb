# <a id ="norm" > Normalization of the vector(v) </a>
    
    u= v/|v|
    
Ex: v= [1/sqrt(2),0,0]
    
    |v|=sqrt(x ^ 2, y ^ 2, z ^ 2)
    
       =sqrt([(1/sqrt(2))^2,0,0])
       
       =sqrt(1/2)
       
       =1/sqrt(2)
    
u= [1/sqrt(2),0,0]/1/sqrt(2)

 = [1,0,0]

```python
import numpy as np

v=[1/sqrt(2),0,0]

normalized_v = v/np.linalg.norm(v)

print(normalized_v)
```

```python
plot_bloch_vector(v)
```

```python
v=[1/sqrt(2),-1/sqrt(2),0]

normalized_v = v/np.linalg.norm(v)

print(normalized_v)
```

```python
plot_bloch_vector(v)
```

# <a id="spherical" >shperical coordinates</a>

```python
plot_bloch_vector([1,pi/2,pi/2],coord_type="spherical")
```

- r, theta, phi = x,y,z
- x = r * np.sin(theta) * np.cos(phi)
- y = r * np.sin(theta) * np.sin(phi)
- z = r * np.cos(theta)

```python
r, theta, phi=1,pi/2,pi/2

x = r * np.sin(theta) * np.cos(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(theta)
```

```python
x,y,z
```

```python
plot_bloch_vector([x,y,z])
```
