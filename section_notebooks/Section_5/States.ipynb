# Statevector class

```python
from qiskit.quantum_info import *
from qiskit import QuantumCircuit
```

### from_label from the statvector class

```python
qc=QuantumCircuit(2)
sv=Statevector.from_label('01')
sv_evolve=sv.evolve(qc)
sv_evolve
```

```python
sv_evolve.data
```

### draw() method of the statevector class from Quantum_info

Reference:
    
https://qiskit.org/documentation/locale/de_DE/stubs/qiskit.quantum_info.Statevector.draw.html#qiskit.quantum_info.Statevector.draw

```python
sv_evolve.draw(output="latex")
```

```python
sv_evolve.draw('qsphere')
```

### from_int from the statvector class

https://qiskit.org/documentation/locale/de_DE/stubs/qiskit.quantum_info.Statevector.from_int.html#qiskit.quantum_info.Statevector.from_int

```python
qc=QuantumCircuit(2)
sv=Statevector.from_int(2,2**2)
sv_evolve=sv.evolve(qc)
sv_evolve
```

```python
sv_evolve.data
```

```python
sv_evolve.draw(output="latex")
```

# Density Matrix class:

Another method to display the above qunatum state is in the form of a matrix.

- Density operator 
- outer product

```python
qc=QuantumCircuit(2)
density_M=DensityMatrix.from_label('01')
density_M=density_M.evolve(qc)
density_M
```

```python
density_M.data
```

```python
density_M.draw(output="latex")
```

```python
density_M.to_operator()
```

```python
density_M.to_statevector()
```

```python
density_M.draw('city')
```
