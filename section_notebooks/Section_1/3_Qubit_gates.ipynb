# Toffoli gate
***

### 3-qubit gate
SYNTAX: ccx(control,control,target)
    
   https://qiskit.org/documentation/stubs/qiskit.circuit.library.CCXGate.html#qiskit.circuit.library.CCXGate

```python
from qiskit import *
from qiskit.quantum_info import Statevector
from math import pi
```

```python
qc=QuantumCircuit(3)
qc.ccx(0,1,2)
qc.draw(output="mpl")
```

```python
backend = Aer.get_backend('unitary_simulator')

job = execute(qc, backend)
result = job.result()
#Obtain the unitary matrix for ccx gate
print(result.get_unitary(qc, decimals=3))
```

```python
state = Statevector.from_instruction(qc)
state.draw(output='latex')
```


Reference: IBM Quantum Composer

### Target qubit is flipped only when the first 2 control qubits are in |1> state

```python
qc=QuantumCircuit(3)
qc.x(0)
qc.x(1)
qc.ccx(0,1,2)
qc.draw(output="mpl")
```

```python
state = Statevector.from_instruction(qc)
state.draw(output='latex')
```
