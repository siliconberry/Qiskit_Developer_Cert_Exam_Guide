```python
from qiskit import *
from qiskit.quantum_info import Statevector
from math import pi
```

# Rotation Gates: RX, RY and RZ
# Are Non-Clifford gates

***
# Rx gate:
***

-Rotation around x-axis

-single-qubit rotation through angle Î¸(radians) around the x-axis

```python
qc=QuantumCircuit(1)
qc.rx(pi/2,0)
qc.draw(output="mpl")
```

```python
backend = Aer.get_backend('unitary_simulator')

job = execute(qc, backend)
result = job.result()
#Obtain the unitary matrix for Rx gate
print(result.get_unitary(qc, decimals=3))
```

```python
state = Statevector.from_instruction(qc)
state.draw(output='latex')
```


Reference: IBM Quantum Composer

### If the qubit is in state |1>

```python
qc=QuantumCircuit(1)
qc.x(0)
qc.rx(pi/2,0)
qc.draw(output="mpl")
```
