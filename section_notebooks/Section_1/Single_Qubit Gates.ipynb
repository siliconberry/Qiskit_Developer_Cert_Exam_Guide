***
> # Pauli-X gate
***

### Also called as "NOT gate"-Classical Analagous

### Called as "bit-flip" gate

https://qiskit.org/documentation/stubs/qiskit.circuit.library.XGate.html#qiskit.circuit.library.XGate


```python
qc=QuantumCircuit(1)
qc.x(0)
qc.draw(output="mpl")
```

##  To understand about unitary refer to Section 6</a>

```python
backend = Aer.get_backend('unitary_simulator')
job = execute(qc, backend)
result = job.result()
#Obtain the unitary matrix for Pauli-X gate
print(result.get_unitary(qc, decimals=3))
```




Reference: IBM Quantum Composer

***
> # P gate(theta)
***

https://qiskit.org/documentation/stubs/qiskit.circuit.library.PhaseGate.html#qiskit.circuit.library.PhaseGate

## Phase gate
## Parametrized
## rotates by an angle theta around Z-axis


```python
qc=QuantumCircuit(1)
qc.p(pi,0)

qc.draw(output="mpl")
```

## Theta=pi
## We obtain Z gate

```python
backend = Aer.get_backend('unitary_simulator')

job = execute(qc, backend)
result = job.result()
#Obtain the unitary matrix for P gate
print(result.get_unitary(qc, decimals=3))
```

```python
qc=QuantumCircuit(1)
qc.p(pi/2,0)

qc.draw(output="mpl")
```

## Theta=pi/2
## We obtain S gate

```python
backend = Aer.get_backend('unitary_simulator')

job = execute(qc, backend)
result = job.result()
#Obtain the unitary matrix for P gate
print(result.get_unitary(qc, decimals=3))
```

```python
qc=QuantumCircuit(1)
qc.p(pi/4,0)

qc.draw(output="mpl")
```

## Theta=pi/4
## We obtain T gate

```python
backend = Aer.get_backend('unitary_simulator')

job = execute(qc, backend)
result = job.result()
#Obtain the unitary matrix for P gate
print(result.get_unitary(qc, decimals=3))
```
