# Controlled NOT Gate : CX
***
### Also called as controlled-x gate
### conditional gate
### Syntax: cx(control,target)
https://qiskit.org/documentation/stubs/qiskit.circuit.library.CXGate.html#qiskit.circuit.library.CXGate

### CASE 1: q0 is control, q1 is target

```python
qc=QuantumCircuit(2)
qc.cx(0,1)
qc.draw(output="mpl")
```

```python
backend = Aer.get_backend('unitary_simulator')

job = execute(qc, backend)
result = job.result()
#Obtain the unitary matrix for cx gate
print(result.get_unitary(qc, decimals=3))
```

### CASE 2: q1 is control, q0 is target

```python
qc=QuantumCircuit(2)
qc.cx(1,0)
qc.draw(output="mpl")
```

```python
backend = Aer.get_backend('unitary_simulator')

job = execute(qc, backend)
result = job.result()
#Obtain the unitary matrix for cx gate
print(result.get_unitary(qc, decimals=3))
```

### Operation: cx gate flips the target qubit if the control qubit is in |1> state

```python
qc=QuantumCircuit(2)
qc.cx(0,1)
qc.draw(output="mpl")
```

```python
state = Statevector.from_instruction(qc)
```

```python
print(state)
```

```python
state.draw(output='latex')
```
