# decompose ()

Call a decomposition pass on this circuit, to decompose one level (shallow decompose).


Reference:
    
https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.decompose.html

```python
from qiskit import QuantumCircuit
import numpy as np
```

# Example 1: toffoli gate

```python
qc = QuantumCircuit(3)
qc.ccx(0,1,2)
qc.draw(output="mpl")
```

```python
qc.decompose().draw(output="mpl")
```

# Example 2: single qubit gate and 2-qubit gate

```python
qc = QuantumCircuit(2)
qc.y(0)
qc.z(1)
qc.h(0)
qc.swap(0,1)
qc.draw(output="mpl")
```

```python
qc.decompose().draw(output="mpl")
```

```python
qc.decompose().decompose().draw(output="mpl")
```

# Example 3: let's take a circuit with initialize()

```python
qc = QuantumCircuit(1)
qc.initialize([1/np.sqrt(2), -1/np.sqrt(2)], 0)
qc.draw(output="mpl")
```

```python
qc.decompose().draw(output="mpl")
```

```python
qc.decompose().decompose().draw(output="mpl")
```

```python
qc.decompose().decompose().decompose().draw(output="mpl")
```

```python
qc.decompose().decompose().decompose().decompose().draw(output="mpl")
```
